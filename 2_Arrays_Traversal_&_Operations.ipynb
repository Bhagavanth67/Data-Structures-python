```python
#Write a program to print all elements of an array in forward and reverse order using loops. Example: arr = [1, 3, 5, 7] Forward → 1 3 5 7 Reverse → 7 5 3 1
arr = [1, 3, 5, 7]
print("Forward →", end=" ")
for i in range(len(arr)):
    print(arr[i], end=" ")
print()
print("Reverse →", end=" ")
for j in range(len(arr) - 1, -1, -1):
    print(arr[j], end=" ")
```

```python
#Write a program to find the second largest element in an array. Example: arr = [12, 35, 1, 10, 34, 1] → Output: 34
arr = [12, 35, 1, 10, 34, 1]
maxi = float('-inf')
sec_max = float('-inf')
for i in arr:
    if i > maxi :
        maxi = i
    if i > sec_max and i < maxi:
        sec_max = i
print(f"Output : {sec_max}")
```

```python
# Write a program to insert an element at a specific position in an array. Example: arr = [1, 2, 4, 5], insert 3 at position 2 → Output: [1, 2, 3, 4, 5]
arr = [1, 2, 4, 5]
pos = 2
num = 3
new_arr = []
l = len(arr) + 1
for i in range(l):
    if i < pos:
        new_arr.append(arr[i])
    elif i == pos:
        new_arr.append(num)
    else:
        new_arr.append(arr[i - 1])
print(new_arr)
```

```python
#Write a program to delete an element from an array by its value.
#Example: arr = [10, 20, 30, 40, 50], delete 30 → Output: [10, 20, 40, 50]
arr = [10, 20, 30, 40, 50]
arr = [i for i in arr if i != 30]
print(arr)
```

```python
#Write a program to rotate an array to the left by one position. Example: arr = [1, 2, 3, 4, 5] → Output: [2, 3, 4, 5, 1]
arr = [1, 2, 3, 4, 5]
first = arr[0]
for i in range(len(arr) - 1):
    arr[i] = arr[i + 1]
arr[-1] = first
print(arr)
```
