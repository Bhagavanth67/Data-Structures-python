```python
# Q1: Bubble Sort – Sort an array in ascending order.arr = [5, 1, 4, 2, 8]
arr = [5, 1, 4, 2, 8]
for i in range(0,len(arr)-1):
    for j in range(len(arr) - i - 1):
        if arr[j] > arr[j+1]:
            arr[j] , arr[j+1] = arr[j+1] , arr[j]
print(f"Sorted array : ",arr)
```

```python
# Q2: Selection Sort – Sort an array in ascending order. arr = [64, 25, 12, 22, 11]
arr = [64, 25, 12, 22, 11]
for i in range(len(arr)):
    min_index = i
    for j in range(i + 1, len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
print("Sorted array:", arr)
```

```python
# Q3: Insertion Sort – Sort an array in ascending order. arr = [12, 11, 13, 5, 6]
arr = [12, 11, 13, 5, 6]
for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
print("Sorted array:", arr)
```

```python
# Q4: Sort an array in descending order using Bubble Sort. arr = [3, 9, 1, 4, 7]
arr = [3, 9, 1, 4, 7]
for i in range(0,len(arr)-1):
    for j in range(len(arr) - i - 1):
        if arr[j] < arr[j+1]:
            arr[j] , arr[j+1] = arr[j+1] , arr[j]
print(f"Sorted array : ",arr)
```

```python
# Q5: Given two sorted arrays, merge them into a single sorted array (without using sort()). arr1 = [1, 3, 5, 7], arr2 = [2, 4, 6, 8]
arr1 = [1, 3, 5, 7]
arr2 = [2, 4, 6, 8]
i = j = 0
merged = []

while i < len(arr1) and j < len(arr2):
    merged.append(arr1[i] if arr1[i] < arr2[j] else arr2[j])
    if arr1[i] < arr2[j]: i += 1
    else: j += 1

merged += arr1[i:] + arr2[j:]
print("Merged array:", merged)
```
