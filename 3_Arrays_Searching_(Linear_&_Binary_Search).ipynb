```python
# Q1: Linear Search = arr = [10, 20, 30, 40, 50], key = 30
arr = [10, 20, 30, 40, 50]
key = 30
found = False

for i in range(len(arr)):
    if arr[i] == key:
        print("Element found at index:", i)
        found = True
        break

if not found:
    print("Element not found")
```

```python
# Q2: Binary Search (Iterative) arr = [2, 4, 6, 8, 10, 12], key = 10
arr = [2, 4, 6, 8, 10, 12]
key = 10
low = 0
high = len(arr) - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print("Element found at index:", mid)
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print("Element not found")
```

```python
# Q3: Binary Search (Recursive) arr = [5, 15, 25, 35, 45], key = 35
def binary_search(arr, low, high, key):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == key:
        return mid
    elif arr[mid] < key:
        return binary_search(arr, mid + 1, high, key)
    else:
        return binary_search(arr, low, mid - 1)
arr = [5, 15, 25, 35, 45]
key = 35
result = binary_search(arr, 0, len(arr) - 1, key)
if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

```python
# Q4: First & Last Occurrence (Binary Search Logic) arr = [2, 4, 4, 4, 6, 8], key = 4
def first_occurrence(arr, key):
    low = 0
    high = len(arr) - 1
    first = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            first = mid
            high = mid - 1   # keep searching left
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return first
def last_occurrence(arr, key):
    low = 0
    high = len(arr) - 1
    last = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            last = mid
            low = mid + 1    # keep searching right
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return last
arr = [2, 4, 4, 4, 6, 8]
key = 4
first = first_occurrence(arr, key)
last = last_occurrence(arr, key)
if first != -1:
    print("First occurrence at index:", first)
    print("Last occurrence at index:", last)
else:
    print("Element not found")
```

```python
# Q5: Count Occurrences (Binary Search Based) arr = [1, 2, 2, 2, 3, 4, 5], key = 2
def first_occurrence(arr, key):
    low = 0
    high = len(arr) - 1
    first = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            first = mid
            high = mid - 1
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return first
def last_occurrence(arr, key):
    low = 0
    high = len(arr) - 1
    last = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            last = mid
            low = mid + 1
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return last
arr = [1, 2, 2, 2, 3, 4, 5]
key = 2
first = first_occurrence(arr, key)
last = last_occurrence(arr, key)
if first == -1:
    print("Element not found")
else:
    count = (last - first) + 1
    print("Count of", key, "is:", count)
```
