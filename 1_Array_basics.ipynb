```python
#Write a program to find the sum and average of all elements in an array
arr = [2, 5, 8, 10]
total = 0
for num in arr:
    total += num
average = total / len(arr)
print("Sum of all elements in an array is", total)
print("Average of all elements in an array is", average)
```

```python
#Write a program to find the maximum and minimum elements in an array without using built-in functions. Example: arr = [7, 2, 9, 4, 1] → Output: Max = 9, Min = 1
arr = [7, 2, 9, 4, 1]
maxi = arr[0]
mini = arr[0]
for i in arr:
    if i < mini:
        mini = i
    elif i > maxi:
        maxi = i
print("Maximum value of array is ",maxi)
print("Minimum value of array is ",mini)
```

```python
#Write a program to reverse an array in place (without using extra space). Example: arr = [1, 2, 3, 4, 5] → Output: [5, 4, 3, 2, 1]
arr = [1, 2, 3, 4, 5]
for i in range(len(arr)//2):
    arr[i], arr[len(arr)-i-1] = arr[len(arr)-i-1], arr[i]
print(arr)
```

```python
#Write a program to count the number of positive, negative, and zero elements in an array. Example: arr = [3, -1, 0, 4, -2, 0] → Output: Positive = 2, Negative = 2, Zero = 2
arr = [3, -1, 0, 4, -2, 0]
Positive = 0
Negative = 0
Zero = 0
for i in arr:
    if i > 0:
        Positive += 1
    elif i == 0:
        Zero += 1
    else:
        Negative += 1
print("Positive =",Positive)
print("Negative =",Negative)
print("Zero =",Zero)
```

```python
#Write a program to copy all elements from one array to another without using slicing or library functions. Example: Input → [10, 20, 30] → Output: Copied array = [10, 20, 30]
arr = [10, 20, 30]
arr_new = []
for i in arr:
    arr_new.append(i)
print("Copied array =",arr_new)
```
